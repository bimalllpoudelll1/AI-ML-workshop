```python
# matplotlib and seaborn --- IGNORE ---
import matplotlib.pyplot as plt
fruits=['apple','banana','orange','grape']
sales=[100,150,80,120]
revenue=[200,300,160,240]
```

```python
plt.Figure(figsize=(10,5))


plt.figure(figsize=(6,4))
plt.plot(fruits, sales, c='red', marker='o')
plt.plot(fruits, revenue, c='green', marker='p')
plt.title("Fruit sales and revenue")
plt.xlabel("total sales")
plt.ylabel("fruit")
plt.show()
```
